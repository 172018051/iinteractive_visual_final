```python
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Bar, Page, Pie, Timeline
from pyecharts.charts import Map

import pandas as pd
from pandas import DataFrame
import numpy as np

#准备
data = pd.read_csv('分省年度数据，粗离婚率%.csv',engine='python') #
data
D = list(data.地区)

def timeline_map() -> Timeline:
    tl = Timeline()
    for i in range(2010, 2018):
        map0 = (
            Map()
            .add(
                "粗离婚率",list(zip(list(D),list(data["{}年".format(i)]))),"china",is_map_symbol_show = False
            )
            .set_global_opts(
                title_opts=opts.TitleOpts(title="中国各省{}年粗离婚率".format(i),subtitle="",
                                          subtitle_textstyle_opts=opts.TextStyleOpts(color="red",font_size=18,font_style="italic")),
                visualmap_opts=opts.VisualMapOpts(min_=0,max_=5),
            
            )
        )
        tl.add(map0,"{}年".format(i))
    return tl

地图 = timeline_map()
地图.render_notebook()
```
